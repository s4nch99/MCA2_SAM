f(x) = \sqrt{-x} \qquad g(x) = \ln x

gráfica
\begin{equation}
f\circ g \qquad g ∘ f \qquad fg

```python
from sympy import symbols, sqrt, ln, plot
x = symbols('x')
f = sqrt(-x)
g = ln(x)
p1 = plot(f.subs(x, g), (x, 0.01, 1), show=False, line_color='red')
p2 = plot(g.subs(x, f), (x, -5, -0.01), show=False, line_color='blue')
p3 = plot(f * g, (x, -5, -0.01), show=False, line_color='green')
p1.show()
p2.show()
p3.show()
```


## Actividad:

Consider the following function.

\begin{equation}
f(x) = \sum_{k=1}^{100} \left( \frac{\sin(2\pi k^2 x)}{4\pi^2k^5} + \frac{x^2}{2k}\right)
\end{equation}
This function was suggested by Sungkon Chang as an example of a function
that looks quite “innocent” but whose derivatives behave quite wildly. Plot 𝑓 (𝑥),
𝑓 ′ (𝑥) and 𝑓 ′′ (𝑥) and observe this behaviour.

```python
from sympy import symbols, sin, pi, Sum, diff, plot
x, k = symbols('x k')
termino = (sin(2 * pi * k**2 * x) / (4 * pi**2 * k**5)) + (x**2 / (2 * k))
f_x = Sum(termino, (k, 1, 100)).doit()
f_prima = diff(f_x, x)
f_biprima = diff(f_prima, x)
plot(f_x, (x, -0.5, 0.5), line_color='blue')
plot(f_prima, (x, -0.5, 0.5), line_color='red')
plot(f_biprima, (x, -0.5, 0.5), line_color='green')
```

## Funciones Paramétricas

```python
from sympy import symbols, cos, sin, exp, pi
from sympy import plot_parametric
t = symbols('t')
def x(t):
  return sin(t) * (exp(cos(t)) - 2 * cos(4 * t) - sin(t/12)**5)
def y(t):
  return cos(t) * (exp(cos(t)) - 2 * cos(4*t) - sin(t/12)**5)

plot_parametric(x(t), y(t), (t, -6*pi, 6*pi));
plot_parametric(x(t), y(t), (t, -50, 50));
```

```python
import numpy as np
import matplotlib.pyplot as plt
theta = np.linspace(0, 2*np.pi, 1000)
valores_a = [1, 2, 3, 5, 6]
plt.figure(figsize=(8,8))
for a in valores_a:
    r2 = a**2 * np.cos(2*theta)
    r = np.sqrt(np.abs(r2))
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    plt.plot(x, y, label=f'a={a}')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

Al cambiar el valor del parámetro, aumenta el tamaño de las curvas (pétalos) de la lemniscata. El parámetro "a" determina la escala a la que se extiende la gráfica desde el origen, manteniendo la forma geométrica pero incrementando su amplitud.

## Lemniscata en coord. polares
$$
r^2 = a^2 \cos(2\theta)
$$
donde
$$
x = r\cos(\theta) \;\; \; \; y = r\sin(\theta)
$$



```python
import numpy as np
import matplotlib.pyplot as plt

a=2
theta = np.linspace(0,  2*np.pi, 1000)
r_cuadrado = a**2*np.cos(2*theta)
r = np.sqrt(np.maximum(r_cuadrado, 0))

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.plot(x,y);
plt.title("Lemniscata de Bernoulli")
plt.grid(True,alpha=0.3)
plt.axis('equal');
```

## Ejemplo: **F**(x,y) = (-y, x)

---

Referencia:
* Campo_Direccional_EDO1erOrden_v3.ipynb

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función M y N
def M(x,y):
    return -y

def N(x,y):
    return x

nx, ny = .2, .2
x = np.arange(-2, 2, nx)
y = np.arange(-2, 2, ny)

# MESHGRID
X, Y = np.meshgrid(x, y)

# Funciones Componente
dx = M(X,Y)
dy = N(X,Y)


# Normalizar
dyu = dy/np.sqrt(dx**2 + dy**2)
dxu = dx/np.sqrt(dx**2 + dy**2)

# Gráficas
plt.quiver(X,Y,dxu,dyu, color = "orange",  headwidth = 2)
#plt.xticks(x, rotation = 60)
#plt.yticks(y)
plt.title("Vector Field F = (-y, x)", color='blue',
              fontsize ='x-large')
plt.show()
```
